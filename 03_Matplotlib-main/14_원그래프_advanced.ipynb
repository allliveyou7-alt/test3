# 14. 원 그래프 심화
---
### 원 그래프 고급 기능
- 색상, 도넛 차트, 조건부 표시 등 다양한 스타일링 옵션을 활용한다.
- DataFrame 데이터를 활용하여 실제 데이터를 시각화한다.


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 사용자
matplotlib.rcParams['font.size'] = 15 # 폰트 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
```

```python
# 온라인 쇼핑몰 카테고리별 매출 비율
categories = ['의류', '전자제품', '화장품', '식품', '도서', '기타']
sales_ratio = [35, 25, 15, 12, 8, 5]  # 매출 비율 (%) 100이 맞지않아도 자동계산됨됨
```

```python
len(categories)
explode=[0.1]*len(categories)


# 파스텔 계열 직접 지정
pastel_colors = [
    '#FFB5E8',  # 핑크
    '#B5EAD7',  # 민트
    '#C7CEEA',  # 연보라
    '#FFDAC1',  # 살구
    '#E2F0CB',  # 연두
    '#FFFFD1'   # 연노랑
]

plt.pie(sales_ratio, labels=categories, autopct='%.1f%%', startangle=90, explode=explode, colors=pastel_colors)
plt.title('제품 카테고리별 매출 비율(파스텔톤)')
plt.show()
```

## 도넛 차트 만들기
### wedgeprops 파라미터
- **wedgeprops**의 **width** 속성으로 도넛 차트를 만들 수 있다.
- width를 0과 1 사이의 값으로 설정하면 중앙에 구멍이 생긴다.


```python
len(categories)
explode=[0.03]*len(categories)


# 파스텔 계열 직접 지정
pastel_colors = [
    '#FFB5E8',  # 핑크
    '#B5EAD7',  # 민트
    '#C7CEEA',  # 연보라
    '#FFDAC1',  # 살구
    '#E2F0CB',  # 연두
    '#FFFFD1'   # 연노랑
]

wedgeprops = {
    'width':0.5,
    'edgecolor' : 'w',
    'linewidth' : 5
    }

plt.pie(sales_ratio, labels=categories, autopct='%.1f%%',
        startangle=90, explode=explode, colors=pastel_colors,
        wedgeprops=wedgeprops)
plt.title('제품별 매출 비율')
plt.show()
```

### 테두리 스타일
- **edgecolor**: 조각 테두리 색상
- **linewidth**: 테두리 선 두께


## 조건부 백분율 표시
### 사용자 정의 autopct 함수
- 특정 조건에 따라 백분율을 선택적으로 표시할 수 있다.
- 예: 10% 이상인 경우에만 표시


```python
len(categories)
explode=[0.03]*len(categories)


# 파스텔 계열 직접 지정
pastel_colors = [
    '#FFB5E8',  # 핑크
    '#B5EAD7',  # 민트
    '#C7CEEA',  # 연보라
    '#FFDAC1',  # 살구
    '#E2F0CB',  # 연두
    '#FFFFD1'   # 연노랑
]

def custom_autopct(pct):
    return ('%.1f%%' % pct) >= 10 else ''

wedgeprops = {
    'width':0.5,
    'edgecolor' : 'w',
    'linewidth' : 5
    }

plt.pie(sales_ratio, labels=categories, autopct='%.1f%%',
        startangle=90, explode=explode, colors=pastel_colors,
        wedgeprops=wedgeprops)
plt.title('제품별 매출 비율')
plt.show()
```
