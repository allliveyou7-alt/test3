# 04. 그래프 스타일 설정
---
### 그래프 시각화 향상
- 그래프의 선 스타일, 색상, 마커 등을 변경하여 더 명확하고 보기 좋은 시각화를 만들 수 있다.
- 여러 데이터를 한 그래프에 표시할 때 각 데이터를 쉽게 구분할 수 있도록 스타일을 설정한다.


```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 사용자
matplotlib.rcParams['font.size'] = 15 # 폰트 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
```

## 선 색상 설정
### color 파라미터
- **color** 파라미터를 사용하여 선의 색상을 변경할 수 있다.
- 색상 지정 방법:
    - 색상 이름: 'red', 'blue', 'green' 등
    - 단축 코드: 'r', 'b', 'g', 'c', 'm', 'y', 'k', 'w'
    - RGB 코드: '#FF0000' (빨강), '#00FF00' (초록) 등


```python
x=[1,2,3]
y=[2,4,8]

plt.plot(x,y, color='red')
plt.title('빨간색 그래프')

plt.show()
```

## 선 스타일 설정
### linestyle (ls) 파라미터
- **linestyle** 또는 **ls** 파라미터로 선의 스타일을 변경할 수 있다.
- 주요 선 스타일:
    - '-' : 실선 (기본값)
    - '--' : 대시선
    - '-.' : 대시-점선
    - ':' : 점선


```python
plt.plot(x,y,ls='-')
plt.plot(x,[i+1 for i in y],ls='--')
plt.plot(x,[i+2 for i in y],ls=':')

plt.title('선스타일 지정')

plt.show()
```

## 마커 설정
### marker 파라미터
- **marker** 파라미터로 데이터 포인트에 표시할 마커를 지정할 수 있다.
- 주요 마커 스타일:
    - 'o' : 원
    - 's' : 사각형
    - '^' : 삼각형 (위)
    - 'v' : 삼각형 (아래)
    - '*' : 별
    - '+' : 플러스
    - 'x' : X 표시


```python
plt.plot(x,y,marker='^')
plt.show()
```

### 마커 크기와 색상
- **markersize (ms)**: 마커의 크기를 조정한다.
- **markerfacecolor (mfc)**: 마커 내부 색상을 지정한다.
- **markeredgecolor (mec)**: 마커 테두리 색상을 지정한다.

> 키워드인자 (ms, mfc, mec )
> - 약어스타일로 설정이 가능하다.


```python
# plt.plot(x,y,marker='v', markersize=15, markerfacecolor='red', markeredgecolor='blue')
plt.plot(x,y,marker='v', ms=15, mfc='red', mec='blue')
plt.show()
```

## 선 두께 설정
### linewidth (lw) 파라미터
- **linewidth** 또는 **lw** 파라미터로 선의 두께를 조정할 수 있다.
- 기본값은 1.5이며, 숫자가 클수록 선이 두꺼워진다.


```python
plt.plot(x,y,lw=10)
```

## 여러 데이터 시각화
### 3회사의 주가를 표로 표현해보자
- 여러 개의 데이터를 한 그래프에 표시하고, 각각을 구분하기 위해 다양한 스타일을 적용한다.


```python
days = [1, 2, 3, 4, 5] # 1일 , 2일, 3일, 4일, 5일
samsung = [70000, 72000, 68000, 71000, 73000] # 삼성전자 주가 (단위: 원)
kakao = [50000, 48000, 52000, 51000, 53000] # 카카오
naver = [200000, 205000, 198000, 210000, 215000] # 네이버


```

```python
import matplotlib.pyplot as plt

# 데이터
days = [1, 2, 3, 4, 5] # 날짜
samsung = [70000, 72000, 68000, 71000, 73000] # 삼성전자 주가
kakao = [50000, 48000, 52000, 51000, 53000] # 카카오 주가
naver = [200000, 205000, 198000, 210000, 215000] # 네이버 주가

plt.figure(figsize=(8,6))

# 꺾은선 차트 그리기
plt.plot(days, samsung, label='삼성', color='blue', linestyle='-')  # 실선
plt.plot(days, kakao, label='카카오', color='orange', linestyle='--')  # 점선
plt.plot(days, naver, label='네이버', color='green', linestyle='-.')  # 대시점선

plt.xlabel('날짜')
plt.ylabel('주가 (원)')
plt.title('주가 현황')
plt.legend()  # 범례
plt.grid(True)  # 격자 표시

plt.xticks(days, [f'{d}일' for d in days])  # x축 레이블 예쁘게

plt.tight_layout()
plt.show()
```
