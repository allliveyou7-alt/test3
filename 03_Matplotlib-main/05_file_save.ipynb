# 05. 그래프 파일로 저장하기
---
### 그래프 저장의 필요성
- 생성한 그래프를 이미지 파일로 저장하여 보고서, 프레젠테이션 등에 활용할 수 있다.
- Matplotlib는 다양한 이미지 형식으로 그래프를 저장할 수 있다.


```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 사용자
matplotlib.rcParams['font.size'] = 15 # 폰트 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
```

```python
# 기본 데이터 준비
x = [1, 2, 3]
y = [2, 4, 8]
```

## savefig() 함수
### 기본 사용법
- **savefig()** 함수를 사용하여 그래프를 이미지 파일로 저장할 수 있다.
- 파일명의 확장자에 따라 자동으로 형식이 결정된다.
- 주요 형식: PNG, JPG, PDF, SVG 등


```python
plt.plot(x,y)
plt.title('기본 그래프')
plt.savefig('image/graph_basic.png')
plt.show()
```

### 이미지 해상도 조정
- **dpi** (dots per inch) 파라미터로 이미지의 해상도를 조정할 수 있다.
- 기본값은 100이며, 값이 클수록 고해상도 이미지가 생성된다.
- 인쇄용이나 고품질 이미지가 필요할 때는 300 이상을 권장한다.


```python
plt.plot(x,y)
plt.title('기본 그래프')
plt.savefig('image/graph_basic_high_quality.png', dpi=200)
plt.show()
```

### 여백 조정
- **bbox_inches** 파라미터를 'tight'로 설정하면 그래프 주변의 불필요한 여백을 제거할 수 있다.
- 공간을 효율적으로 사용할 수 있어 문서에 삽입할 때 유용하다.


```python
plt.plot(x,y)
plt.title('기본 그래프')
plt.savefig('image/graph_basic_tight.png', bbox_inches='tight')
plt.show()
```

### 배경색 설정
- **facecolor** 파라미터로 그래프의 배경색을 지정할 수 있다.
- 투명 배경이 필요한 경우 **transparent=True**를 사용한다.


```python
plt.plot(x,y)
plt.title('기본 그래프')
plt.savefig('image/graph_backgroundcolor.png', facecolor='lightgray')
plt.show()
```

```python
plt.plot(x,y)
plt.title('기본 그래프')
plt.savefig('image/graph_transparent.png', transparent=True)
plt.show()
```
